```python
import yfinance as yf
import datetime

# Choose stock
ticker = "AAPL"   # you can also try "RELIANCE.NS", "TCS.NS", "SBIN.NS"

# Download proper clean data using Ticker().history()
stock = yf.Ticker(ticker)

df = stock.history(period="6mo", interval="1d")   # <â€” THIS RETURNS CLEAN COLUMNS

df = df.reset_index()   # ensures Date becomes a normal column

df.head()
```

```python
import plotly.graph_objects as go

fig = go.Figure(data=[go.Candlestick(
    x=df['Date'],
    open=df['Open'],
    high=df['High'],
    low=df['Low'],
    close=df['Close']
)])

fig.update_layout(
    title=f"{ticker} - Candlestick Chart (Last 6 Months)",
    xaxis_title="Date",
    yaxis_title="Price",
    xaxis_rangeslider_visible=True
)

fig.show()
```
