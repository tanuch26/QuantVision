#### Use NumPy to generate a random number between 0 and 1

```python
import numpy as np
import random

#function defined for a random number
def ran():
  y= random.random()
  if y>0 and y<1:
    return y
  else:
      ran()

#array returning that number
arr=np.array([ran()])

print(arr)
```

#### Use NumPy to generate an array of 25 random numbers sampled from a standard normal distribution

```python
import numpy as np

arr = np.random.normal(0, 1, 25)
print(arr)

```

#### Create the following matrix:

```python
import numpy as np

arr=np.arange(0.01,1.01,0.01)
arr=arr.reshape(10,10)
print(arr)
```

#### Create an array of 20 linearly spaced points between 0 and 1:

```python
import numpy as np

arr=np.arange(0,1.01,0.05263158)
arr[19]=1

print(arr)
```
