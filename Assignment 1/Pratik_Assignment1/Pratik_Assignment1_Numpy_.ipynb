## Numpy Indexing and Selection

Now you will be given a few matrices, and be asked to replicate the resulting matrix outputs:

```python
arr11=np.array(range(1,26))
arr11= arr11.reshape(5,5)
print(arr11)
```

```python
arr12=arr11[2:,1:]
print(arr12)
```

```python
print(arr12[1,-1])
```

```python
arr12=arr11[0:3,1]
arr12=arr12.reshape(3,-1)
print(arr12)
```

```python
arr12=arr11[-1]
arr12=arr12.reshape(1,-1)
print(arr12)
```

```python
arr12=arr11[-2:]
print(arr12)
```
